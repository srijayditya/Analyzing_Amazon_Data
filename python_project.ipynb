```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('python project main.csv', encoding='ISO-8859-1')

# Convert the "Order Date" column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Calculate the total sales, maximum sale, minimum sale, median sale, and mean sale
total_sales = np.sum(df['Total Owed'])
max_sale = np.max(df['Total Owed'])
min_sale = np.min(df['Total Owed'])
median_sale = np.median(df['Total Owed'])
mean_sale = np.mean(df['Total Owed'])

# Print the calculated statistics
print("Total Sales:", total_sales)
print("Maximum Sale:", max_sale)
print("Minimum Sale:", min_sale)
print("Median Sale:", median_sale)
print("Mean Sale:", mean_sale)


# Plot a bar chart of the total orders per day with different color and size
plt.figure(figsize=(20, 10))
plt.bar(df['Order Date'], df['Total Owed'], color='#FF5733', width=0.8)  # Color and width changed
plt.xlabel('Order Date')
plt.ylabel('Total Owed')
plt.title('Total Orders per Day')
plt.xticks(rotation=90)
plt.show()

# Group the data by order date and calculate the sum of the total owed for each day
daily_orders = df.groupby('Order Date')['Total Owed'].sum().reset_index()  # Reset the index

# Plot a bar chart of the daily orders with different color and size
plt.figure(figsize=(20, 10))
plt.bar(daily_orders['Order Date'], daily_orders['Total Owed'], color='#33FF57', width=0.8)  # Color and width changed
plt.xlabel('Order Date')
plt.ylabel('Total Owed')
plt.title('Daily Orders')
plt.xticks(rotation=90)
plt.show()


# Extract the order dates and total sales from the grouped DataFrame
order_dates = daily_orders['Order Date']
total_sales = daily_orders['Total Owed']

# Create a line chart of the total sales over time with different color and size
plt.figure(figsize=(20, 10))
plt.plot(order_dates, total_sales, color='#3357FF', linewidth=2)  # Color and linewidth changed
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.title('Total Sales over Time')
plt.xticks(rotation=90)
plt.show()
```
